```python
import os
import json
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python

def plot_calibration(calibration_points, gaze_tracking_data):
    marker_shapes = ['o', 's', '^', 'x', 'D', '*', '+', 'P', 'h']

    plt.figure(figsize=(8, 6))

    for i, (key, point) in enumerate(calibration_points.items()):
        plt.scatter(point['x'], point['y'], color='black', marker=marker_shapes[i], label=key)

    centers_of_gravity = []

    for key, point in calibration_points.items():
        center_x = point['x'] + 10  
        center_y = point['y'] + 10
        centers_of_gravity.append((center_x, center_y))

    for i, (center_x, center_y) in enumerate(centers_of_gravity):
        plt.scatter(center_x, center_y, color='green', marker='x', label=f'Centre de gravité {i+1}')

    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Points de calibration, suivi du regard et centres de gravité')
    plt.show()
```

```python
folder_path = './data_file_calibration'

with open(os.path.join(folder_path, os.listdir(folder_path)[0]), 'r') as file:
    data = json.load(file)

calibration_points = data["calibPoints"]
gaze_tracking_data = data["data"]

plot_calibration(calibration_points, gaze_tracking_data)
```

```python
def detect_calibration_gaze(calibration_points, gaze_tracking_data, threshold_distance):
    calibration_gaze_events = []

    centers_of_gravity = []
    for key, point in calibration_points.items():
        center_x = point['x'] + 10
        center_y = point['y'] + 10
        centers_of_gravity.append((center_x, center_y))

    for gaze_point in gaze_tracking_data:
        gaze_x, gaze_y = gaze_point['x'], gaze_point['y']
        
        for i, (center_x, center_y) in enumerate(centers_of_gravity):
            distance_to_center = math.sqrt((gaze_x - center_x)**2 + (gaze_y - center_y)**2)
            if distance_to_center <= threshold_distance:
                calibration_gaze_events.append((gaze_point['timestamp'], i+1))

    return calibration_gaze_events

threshold_distance = 30  # Définir une distance seuil appropriée
calibration_gaze_events = detect_calibration_gaze(calibration_points, gaze_tracking_data, threshold_distance)

print("Moments où les points de calibration capturent le regard dans leur centre de gravité :")
for timestamp, point_number in calibration_gaze_events:
    print(f"Temps : {timestamp}, Point de calibration : {point_number}")
```
